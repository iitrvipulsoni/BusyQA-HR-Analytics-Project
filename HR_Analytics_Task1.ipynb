# Identifying and Treating Outliers

```python
#use the describe function to see statistical data of the numerical columns

df_copy.describe()
```

```python
#creating boxplot for 'city_development_index' column to check for outliers

plt.boxplot(df_copy['city_development_index'], vert=False)
plt.title("Detecting outliers using Boxplot")
plt.xlabel('values')
plt.ylabel('city_development_index')
plt.show()
```

```python
#calculate boxplot parameters(Q1, Q3, IQR, Lower bound, Upper bound) to identify outlier values for the 'city_development_index' column

q1 = np.percentile(df_copy['city_development_index'], 25)
q3 = np.percentile(df_copy['city_development_index'], 75)
IQR = q3-q1
lower_bound = q1-(1.5*IQR)
upper_bound = q3+(1.5*IQR)

print('Lower bound is {}'.format(lower_bound))
print('Upper bound is {}'.format(upper_bound))
print('Minimum value in the city_development_index column is {}'.format(df_copy['city_development_index'].min()))
print('Maximum value in the city_development_index column is {}'.format(df_copy['city_development_index'].max()))
```

All the values lesser than the lower bound and higher than the upper bound are considered as outliers. 

```python
#count of no. of rows having outlier values in the 'city_development_column'

print('No. of rows having values lesser than the lower bound are {}'.format(df_copy[df_copy['city_development_index'] < 0.7374999999999998].shape[0])) 
print('No. of rows having values higher than the upper bound are {}'.format(df_copy[df_copy['city_development_index'] > 1.0295].shape[0])) 
```

```python
#checking percentage contribution of the rows having outlier values

1295/9200 * 100
```

As 14.07% of rows has outliers values within the 'city_development_index' column, therefore, these values cannot be ignored during analysis.

```python
#creating boxplot for 'training_hours' column to check for outliers

plt.boxplot(df_copy['training_hours'], vert=False)
plt.title("Detecting outliers using Boxplot")
plt.xlabel('values')
plt.ylabel('training_hours')
plt.show()
```

```python
#calculate boxplot parameters(Q1, Q3, IQR, Lower bound, Upper bound) to identify outlier values for the 'training_hours' column

q1 = np.percentile(df_copy['training_hours'], 25)
q3 = np.percentile(df_copy['training_hours'], 75)
IQR = q3-q1
lower_bound = q1-(1.5*IQR)
upper_bound = q3+(1.5*IQR)

print('Lower bound is {}'.format(lower_bound))
print('Upper bound is {}'.format(upper_bound))
print('Minimum value in the training_hours column is {}'.format(df_copy['training_hours'].min()))
print('Maximum value in the training_hours column is {}'.format(df_copy['training_hours'].max()))
```

```python
#count of no. of rows having outlier values in the 'training_hours'

print('No. of rows having values lesser than the lower bound are {}'.format(df_copy[df_copy['training_hours'] < -74.5].shape[0])) 
print('No. of rows having values higher than the upper bound are {}'.format(df_copy[df_copy['training_hours'] > 185.5].shape[0])) 
```

```python
#checking percentage contribution of the rows having outlier values

478/9200 * 100
```

As 5.19% of rows has outliers values within the 'training_hours' column, therefore, these values cannot be ignored during analysis.
