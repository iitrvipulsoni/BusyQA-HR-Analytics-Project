# Importing Libraries

```python
# import pandas, numpy, matplotlib, and seaborn libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Importing Train Dataset 

```python
# importing cleaned train dataset

df = pd.read_excel("Clean_Data_HR Analytics.xlsx")
df.head() # seeing initial 5 rows
```

```python
# dropping enrollee_id column from the updated train dataset

df = df.drop('enrollee_id', axis = 'columns')
df #updated train dataset
```

```python
# selecting categorical columns from the updated train dataset

df.select_dtypes(np.object)
```

# Converting all categorical columns into numerical columns

```python
# creating dummy variables for all categorical columns

dummy1 = pd.get_dummies(df.city)
dummy2 = pd.get_dummies(df.gender)
dummy3 = pd.get_dummies(df.relevent_experience)
dummy4 = pd.get_dummies(df.enrolled_university)
dummy5 = pd.get_dummies(df.education_level)
dummy6 = pd.get_dummies(df.major_discipline)
dummy7 = pd.get_dummies(df.experience)
dummy8 = pd.get_dummies(df.company_size)
dummy9 = pd.get_dummies(df.company_type)
dummy10 = pd.get_dummies(df.last_new_job)
```

```python
# If using linear regression then we can drop 1 column from the created dummy variable columns

# dummy1 = dummy1.iloc[:,1:]
# dummy2 = dummy2.iloc[:,1:]
# dummy3 = dummy3.iloc[:,1:]
# dummy4 = dummy4.iloc[:,1:]
# dummy5 = dummy5.iloc[:,1:]
# dummy6 = dummy6.iloc[:,1:]
# dummy7 = dummy7.iloc[:,1:]
# dummy8 = dummy8.iloc[:,1:]
# dummy9 = dummy9.iloc[:,1:]
# dummy10 = dummy10.iloc[:,1:]
```

```python
# merging all the created dummy variable columns with the updated train dataset

merged_df = pd.concat([df,dummy1,dummy2,dummy3,dummy4,dummy5,dummy6,dummy7,dummy8,dummy9,dummy10], axis='columns')
merged_df
```

```python
# dropping all the columns having categorical values to replace them with their respective dummy variable columns 

final_df = merged_df.drop(['city','gender','relevent_experience','enrolled_university','education_level','major_discipline','experience','company_size','company_type','last_new_job'], axis='columns')
final_df.head()
```

```python
# updating the datatype of all the columns to float

final_df = final_df.astype(float)
```

```python
# checking the datatype of all the columns

final_df.dtypes
```

# Creating ML models

```python
# Assigning columns(except the target column) to 'X' as the independent variable

X = final_df.drop('target', axis='columns')
X #independent variable
```

```python
# Assigning the target column) to 'y' as the dependent variable

y = final_df.target
y #dependent variable
```

# Splitting the train dataset to train the ML models

```python
# importing train_test_split library to split the train dataset into X_train, X_test, y_train, y_test 

# X_train, y_train will be used to train the ML models
# X_test, y_test will be used to calculate the accuracy of the models

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

1. Random Forest Classifier Model

```python
# importing Random Forest Classifier Model 

from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier() # creating the object for the model
model1.fit(X_train,y_train) # training the model using the X_train and y_train
score1 = model1.score(X_test,y_test) # calculating the accuracy of the model using the X_test and y_test
```

2. Decision Tree Classifier Model

```python
# importing Decision Tree Classifier Model

from sklearn.tree import DecisionTreeClassifier
model2 = DecisionTreeClassifier() # creating the object for the model
model2.fit(X_train,y_train) # training the model using the X_train and y_train
score2 = model2.score(X_test,y_test) # calculating the accuracy of the model using the X_test and y_test
```

3. Gaussian NB Model

```python
# importing Gaussian NB Model

from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB() # creating the object for the model
model3.fit(X_train,y_train) # training the model using the X_train and y_train
score3 = model3.score(X_test,y_test) # calculating the accuracy of the model using the X_test and y_test
```

# Choosing the best model out of 3 

```python
data = [['RandomForestClassifier', score1], ['DecisionTreeClassifier', score2], ['GaussianNB', score3]]
```

```python
# creating a table to compare the accuracy score of the 3 models

models = pd.DataFrame(data, columns=['Model Name','Accuracy Score'])
models
```

```python
print("Random Forest Classifier Model has the best accuracy score of {}".format(score1))
```

Therefore, we will use this model to predict the 'Target' values for test dataset.

```python
# dropping enrollee_id column from the test dataset

final_test_data = test_data.drop('enrollee_id', axis='columns')
final_test_data # updated test dataset
```

```python
# merging all the created dummy variable columns with the updated test dataset

merged_final_test_data = pd.concat([final_test_data,dummy11,dummy12,dummy13,dummy14,dummy15,dummy16,dummy17,dummy18,dummy19,dummy20], axis='columns')
merged_final_test_data
```

```python
# dropping the columns having categorical values to replace them with their respective dummy variable columns 

merged_final_test_data = merged_final_test_data.drop(['city','gender','relevent_experience','enrolled_university','education_level','major_discipline','experience','company_size','company_type','last_new_job'], axis='columns')
merged_final_test_data.head()
```

```python
# dropping the columns which are not used in training the model

merged_final_test_data = merged_final_test_data.drop(['nan','city_111','city_140','city_171','city_180','city_31','city_48','city_79','city_8','High School','Primary School'], axis='columns')
```

```python
# updating the datatype of all the columns to float

merged_final_test_data = merged_final_test_data.astype(float)
merged_final_test_data.dtypes #checking the datatype 
```

# Predicting output for the test dataset

```python
# predicting the output using the Random Forest Classifier Model

test_data_prediction = model1.predict(merged_final_test_data) # using predict function
test_data_prediction = pd.DataFrame(test_data_prediction) # creating a dataframe of the output column
test_data_prediction.set_axis(['Target'], axis='columns', inplace=True) # assigning the column name
test_data_prediction.reset_index # resetting index of the column
test_data_prediction #Output Target column
```

```python
# merging the output column using concat function

test_data_result = pd.concat(frames,axis='columns')
test_data_result #test dataset with Target column
```
